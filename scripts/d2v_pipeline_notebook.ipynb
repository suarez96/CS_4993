```python
import pickle
import pandas as pd
from scripts.Embedder import Embedder, Doc2VecEmbedder
from scripts.OccupationPreprocessor import OccupationPreprocessor
import numpy as np
```

```python
d2v_train_df = OccupationPreprocessor.prepare_df(
            './Data/doc2vec_train_set.csv',
            input_column='input',
            code_column='code',
            preprocess_text = True
        )
```

```python
test_df = OccupationPreprocessor.prepare_df(
            './Data/overlap_test_set_v4_acanoc_no_train_data.csv',
            input_column='input',
            code_column='code',
            preprocess_text = True
        )
```

```python
embedder = Doc2VecEmbedder(model_name='trial_11.model', train_data=d2v_train_df, infer_params = {
        'steps':2048,
        'alpha':0.03
    })
```

# TODO, make this work as an apply method for pandas vv

```python
def pipeline(row):
    pred = embedder.infer(row['input'])
    scores_1 = embedder.hyperbolic_scoring(pred, level = 1, topn=10)
    # constraint is prediction of first level
    scores_234 = embedder.hyperbolic_scoring(pred, level = 4, topn=30, level_constraint = scores_1[0][0])
    return scores_234[0][0]
```

```python
sample_size = 100
sample_pipeline_df = test_df.sample(sample_size)
```

```python
embedder.infer(sample_pipeline_df['input'])
```

```python
import time
from sklearn.metrics import accuracy_score
scores = []
for i in range(20):
    sample_pipeline_df = test_df.sample(sample_size)
    start_time = time.time()
    predictions = sample_pipeline_df.apply(pipeline, axis = 1)
    elapsed = time.time()-start_time
    print('time to predict on {0} samples: {1:.2f} seconds ({2:.3f} s/sample)'.format(sample_size,
                                                             elapsed, 
                                                            elapsed/sample_size)
         )
    scores.append(
        accuracy_score(predictions, sample_pipeline_df['code'])
    )
```

```python
import plotly.express as px
px.histogram(scores, nbins=8)
```
